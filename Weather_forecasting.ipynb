```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data=pd.read_csv("/content/DailyDelhiClimateTrain.csv")
print(data.head())
```

```python
fig=px.line(data,x='date',
            y='meantemp',
            title='Mean Temp in Dehli over the years'
)
fig.show()
```

```python
fig2=px.line(data,x='date',
            y='humidity',
            title='Humidity in dehli over the years')
fig2.show()
```

```python
fig3=px.line(data,x='date',
            y='wind_speed',
            title='Wind Speed in Dehli over the years')
fig3.show()
```

```python
fig=px.scatter(data_frame=data,x='humidity',
               y='meantemp',size='meantemp',
               trendline='ols',
               title='Relationship between temp and humidity')
fig.show()
```

```python
data['date']=pd.to_datetime(data['date'],format='%Y-%m-%d')
data['year']=data['date'].dt.year
data['month']=data['date'].dt.month
print(data.head())
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15,10))
plt.title('Temp change in delhi over the years')
sns.lineplot(data=data,x='month',y='meantemp',hue='year')
plt.show()
```

```python
# pip install prophet
```

```python
forecast_data=data.rename(columns={"date":"ds",
                                   "meantemp":"y"})
print(forecast_data)
```

```python
from prophet import Prophet
from prophet.plot import plot_plotly,plot_components_plotly
model =Prophet()

model.fit(forecast_data)
forecasts=model.make_future_dataframe(periods=365)
```

```python
predictions=model.predict(forecasts)
plot_plotly(model,predictions)
```

*Hence by analysing the previous weather data,prediction of next year can be made.*
