# Family_type
- 1 alone
- 2-4 small family
- 5-11 large family
  
  **Conclusions**
- if you are with small family your chances of surviving is high.

```python
def transform_family(num):
  if num==1:
   return 'alone'
  elif num>=2 and num<5:
    return 'small-family'
  else:
    return 'Large-family'
```

```python
df['Family-type']=df['Family size'].apply(transform_family)
```

```python
df['Family-type'].describe()
```

```python
pd.crosstab(df['Survived'],df['Family-type'],normalize='columns')*100
```
